```octave
fname = "MM_PositionList.pos"
json_txt = fileread(fname)
json = jsondecode(json_txt)
position_num = length(json.map.StagePositions.array)
```

```octave
position_list = [];

for pos_idx=1:position_num
    position = struct();
    position.id = pos_idx;
    position.well_name = "";
    position.name = "";
    position.ch = "";
    device_num = length(json.map.StagePositions.array(pos_idx).DevicePositions.array);
    for device_idx=1:device_num
        device_name = json.map.StagePositions.array(pos_idx).DevicePositions.array(device_idx).Device.scalar;
        if strcmp(device_name, 'ZDrive')
            position.z_um = json.map.StagePositions.array(pos_idx).DevicePositions.array(device_idx).Position_um.array;
        end
        if strcmp(device_name, 'PFSOffset')
            position.pfs_um = json.map.StagePositions.array(pos_idx).DevicePositions.array(device_idx).Position_um.array;
        end
        if strcmp(device_name, 'XYStage')
            xy_um = json.map.StagePositions.array(pos_idx).DevicePositions.array(device_idx).Position_um.array;
            position.x_um = xy_um(1);
            position.y_um = xy_um(2);
        end
    end
    position_list =[position_list; position];
end
```
